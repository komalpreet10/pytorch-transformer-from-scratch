```python
from src.data_utils import get_tokenizer, load_train_data, load_test_data

tokenizer  = get_tokenizer()
train_data = load_train_data(tokenizer, sample=1000)
test_data  = load_test_data(sample=100)
```

```python
print("TRAINING DATA")
print(f"Rows    : {len(train_data)}")
print(f"Columns : {train_data.column_names}")

print("\nTEST DATA")
print(f"Rows    : {len(test_data)}")
print(f"Columns : {test_data.column_names}")
```

```python
print("SAMPLE TRAINING ROW")
print(f"Instruction : {train_data[0]['instruction']}")
print(f"Input       : {train_data[0]['input'][:300]}")
print(f"Output      : {train_data[0]['output'][:300]}")
print("\nFORMATTED PROMPT")
print(train_data[0]['text'][:500])
```

```python
import matplotlib.pyplot as plt

input_lengths  = [len(r['input'].split())  for r in train_data]
output_lengths = [len(r['output'].split()) for r in train_data]

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].hist(input_lengths,  bins=50, color="steelblue")
axes[0].set_title("Input Length Distribution (words)")
axes[1].hist(output_lengths, bins=50, color="darkorange")
axes[1].set_title("Output Length Distribution (words)")
plt.tight_layout()
plt.savefig("plots/length_distribution.png")
plt.show()

print(f"Avg input length  : {sum(input_lengths)/len(input_lengths):.0f} words")
print(f"Avg output length : {sum(output_lengths)/len(output_lengths):.0f} words")
```
